2.Write a NumPy program to create a 10x10 matrix, in which the elements on the borders will be equal to 1, and inside 0.

```python
arr2 = np.zeros((10, 10), dtype=int)

arr2[[0, -1], :] = 1
arr2[:, [0, -1]] = 1
arr2
```

3. Write a NumPy program to create a 5x10 matrix of distinct numbers from 1 to 50 and swap two columns in a 2d NumPy array.

```python
import numpy as np

arr3 = np.arange(1, 51).reshape(5, 10)
print("original arr", arr3, "\n")
# swap column 1 and column 3
arr3[:, [1, 3]] = arr3[:, [3, 1]]

arr3
```

4. Write a NumPy program to create a 5x10 matrix of distinct numbers from 1 to 50 and swap two rows in a 2d NumPy array.

```python
arr4 = np.arange(1, 51).reshape(5, 10)
print("original arr", arr4, "\n")

# swap row 0 and row 4
arr4[[0, 4], :] = arr4[[4, 0], :]
arr4
```

10. Create a 10x10 array with random values and find the minimum and maximum values, row-wise and column-wis

```python
a = np.random.randint(0,100,(10, 10))

print("Row-wise minimum:", a.min(axis=1))
print("Row-wise maximum:", a.max(axis=1))

print("Column-wise minimum:", a.min(axis=0))
print("Column-wise maximum:", a.max(axis=0))
```
