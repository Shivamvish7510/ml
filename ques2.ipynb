a. Compute the sum, mean, standard deviation, and variance of a two dimensional random integer array  along the second axis.

```python
A = np.random.randint(1, 100, size=(3, 4))
print("Array:\n", A)

# along second axis (axis = 1 â†’ row-wise)
print("Sum:", np.sum(A, axis=1))
print("Mean:", np.mean(A, axis=1))
print("Standard Deviation:", np.std(A, axis=1))
print("Variance:", np.var(A, axis=1))
```

    b. Get the indices of the sorted elements of a given array.
             B = [56, 48, 22, 41, 78, 91, 24, 46, 8, 33]

```python
B = np.array([56, 48, 22, 41, 78, 91, 24, 46, 8, 33])

sorted_indices = np.argsort(B)

print("Original array:", B)
print("Indices of sorted elements:", sorted_indices)
```

```python
c. Create a 2-dimensional array of size m x n integer elements, also print the shape, type  and data type of the array and 
then reshape it into nx m array, n and m are user inputs given at the  run time.
```

```python
m = int(input("Enter number of rows (m): "))
n = int(input("Enter number of columns (n): "))


arr = np.random.randint(1, 50, size=(m, n))

print("Array:\n", arr)
print("Shape:", arr.shape)
print("Type:", type(arr))
print("Data type:", arr.dtype)

reshaped = arr.reshape(n, m)
print("Reshaped array (n x m):\n", reshaped)
```

d. Test whether the elements of a given array are zero, non-zero
and NaN. Record  the indices of these elements in three separate arrays.

```python
arr = np.array([0, 5, np.nan, 8, 0, 3, np.nan])
print ("array: \n" ,arr)

zero_indices = np.where(arr == 0)
nonzero_indices = np.where(arr != 0)
nan_indices = np.where(np.isnan(arr))

print("Zero indices:", zero_indices)
print("Non-zero indices:", nonzero_indices)
print("NaN indices:", nan_indices)
```
