```python
import pandas as pd
import numpy as np

df = pd.read_csv("Iris.csv")

print("Original DataFrame:\n", df.head())
```

##    a. Identify and count missing values in a dataframe.

```python
print("Missing values in each column:\n", df.isnull().sum())

total_missing = df.isnull().sum().sum()
print("Total missing values:", total_missing)
```

##  b. Drop the column having more than 5 null values.

```python
df = df.loc[:, df.isnull().sum() <= 5]
df
```

## c. Identify the row label having maximum of the sum of all values in a row and drop that row.

```python
numeric_df = df.select_dtypes(include=np.number)

row_to_drop = numeric_df.sum(axis=1).idxmax()
df = df.drop(row_to_drop)

print("Dropped row label:", row_to_drop)
```

##        d. Sort the dataframe on the basis of the first column.

```python
second_col = df.columns[1]
df = df.sort_values(by=second_col)
df
```

##   e. Remove all duplicates from the first column.

```python
df = df.drop_duplicates(subset=second_col)
df
```

##   f. Find the correlation between first and second column and covariance between second and third column

```python
col1 = df.columns[0]
col2 = df.columns[1]
col3 = df.columns[2]

correlation = df[col1].corr(df[col2])
covariance = df[col2].cov(df[col3])

print("Correlation between first & second column:", correlation)
print("Covariance between second & third column:", covariance)
```

##  g. Detect the outliers and remove the rows having outliers.

##      h. Discretize second column and create 5 bins

```python
second_col = df.columns[1]


df['Binned'] = pd.cut(df[second_col], bins=5, labels=False)

print(df[['Binned']].head())
```
