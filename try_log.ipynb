```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import KFold
from sklearn import metrics

```

```python
df = pd.read_csv ("iris.csv")
df.sample(5)
```

```python
df=df[df['Species'].isin(['Iris-setosa','Iris-versicolor','Iris-virginica'])]
```

```python
df['Y']=df['Species'].map({
    'Iris-setosa':0,
    'Iris-versicolor':1,
    'Iris-virginica':2
})

    
```

```python
X = df[['SepalLengthCm','SepalWidthCm']].values
Y = df['Y'].values
```

```python
Y
```

```python
sc=StandardScaler()

X=sc.fit_transform(X)
```

```python
def evaluate_model(model,X,Y):
    kf=KFold(n_splits=5,shuffle=True,random_state=2)
    acc_test_score=[]
    acc_train_score=[]
    for train_index,test_index in kf.split(X):
        X_train,X_test=X[train_index],X[test_index]
        Y_train,Y_test=Y[train_index],Y[test_index]
        model.fit(X_train,Y_train)
        
        Y_testpred=model.predict(X_test)
        Y_trainpred=model.predict(X_train)
        
        acc_test_score.append(metrics.accuracy_score(Y_test,Y_testpred))
        acc_train_score.append(metrics.accuracy_score(Y_train,Y_trainpred))
        
    
        
    return np.mean(acc_test_score),np.mean(acc_train_score)
```

```python
model = LogisticRegression()
print(evaluate_model(model,X,Y))
```
